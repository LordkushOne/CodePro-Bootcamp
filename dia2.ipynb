```python
# Temperatura en exterior

temperatura = float(input("Ingrese la temperatura en grados celsius"))
print(f"La temperatura es de: {temperatura}°C")

#INTERVALO ENTRE -10 Y 18
if -10<=temperatura<=18:
    print("fresquete")

#INTERVALO ENTRE 19 Y 30
elif 19<=temperatura<=30:
    print("Calor'i")
    
#INTERVALO ENTRE 31 Y 45
elif 31<=temperatura<=45:
    print("hakuuuu")
else:
    print("Ñamano mba'e!") 
```
